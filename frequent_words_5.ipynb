```python
def tokenize_ja(text):
    node = tagger.parseToNode(str(text)) #分かち書きしたのが入っている。イテレータ
    while node:
        surface = node.surface.lower() #surfaceは単語の見た目　例：トろ
        feature = node.feature.split(",")
        hinshi = feature[0]
        #genkei = feature[6]
        if surface != "" and hinshi in ["名詞","形容詞", "動詞", "副詞", "助動詞", "接続詞", "連体詞", "感動詞", "接頭詞"]:
            yield surface
        node = node.next
```

```python
def tokenize_ja_2(text, w_class):
    node = tagger.parseToNode(str(text)) #分かち書きしたのが入っている。イテレータ
    while node:
        surface = node.surface.lower() #surfaceは単語の見た目　例：トろ
        feature = node.feature.split(",")
        hinshi = feature[0]
        #genkei = feature[6]
        if w_class == "名詞" and surface != "" and hinshi == "名詞":
            yield surface
        elif w_class == "形容詞" and surface != "" and hinshi == "形容詞":
            yield surface
        elif w_class == "動詞" and surface != "" and hinshi == "動詞":
            yield surface
        node = node.next
```

```python
def tokenize(content):
    return [token for token in tokenize_ja(content)]
```

```python
def tokenize_2(content, w_class):
    return [token for token in tokenize_ja_2(content, w_class)]
```

```python
def frequent(path):
    f_words_list = []
    df = pd.read_csv(path)
    if "lunch" in path:
        for i in df["lunch_review"]:
            txt = tokenize(i)
            for t in txt:
                f_words_list.append(t)
    elif "dinner" in path:
        for i in df["dinner_review"]:
            txt = tokenize(i)
            for t in txt:
                f_words_list.append(t)
    return len(f_words_list)
```

```python
def frequent_2(path, w_class):
    f_words_list = []
    f_words = []
    f_words_count = []
    review_list = [] #レビューごとに全単語を収納
    df = pd.read_csv(path)
    if "lunch" in path:
        for i in df["lunch_review"]:
            txt = tokenize_2(i, w_class)
            review_list.append(txt)
            for t in txt:
                f_words_list.append(t)
    elif "dinner" in path:
        for i in df["dinner_review"]:
            txt = tokenize_2(i, w_class)
            review_list.append(txt)
            for t in txt:
                f_words_list.append(t)
    counter = Counter(f_words_list)
    for word, count in counter.most_common():
        f_words.append(word)
        f_words_count.append(count)
    return f_words, f_words_count, review_list
```

```python
def output(path):
    store_name = path.replace("../pretest_", "").replace(".csv", "")
    w_class_list = ["名詞", "形容詞", "動詞"]
    all_words_time = frequent(dinner_files[0]) #その店の全単語数
    for w_class in w_class_list:
        f_words, f_words_count, review_list = frequent_2(path, w_class)
        #f_words　単語の原型
        #f_words_count f_wordsの登場回数
        #review_list 各レビューに含まれる単語
        a = []
        for i in range(len(f_words)):
            s = 0
            for review in review_list:
                if f_words[i] in review:
                    s += 1
            a.append([f_words[i], 
                      f_words_count[i], 
                      int(10000*f_words_count[i]/int(all_words_time))/100,
                      int(100*s/len(review_list))/100, 
                      int(100*f_words_count[i]/len(review_list))/100, 
                      int(100*f_words_count[i]/s)/100])
            df = pd.DataFrame(a, columns = ["頻出単語",
                                            "登場回数",
                                            "登場回数/全単語数(" + str(all_words_time) + ")%",
                                           "登場レビュー率 /(" +  str(len(review_list)) + "本)",
                                           "レビューあたりの登場回数 /回",
                                           "登場したレビューにおける平均登場回数"])
            df.to_csv(store_name + w_class + ".csv", index=False)            
```

```python
output("../pretest_aozora_dinner.csv")
```
