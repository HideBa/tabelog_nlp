```python
store_name = ["sushitsuu",
             "sushiichi",
             "sawada",
             "mitani",
             "matsukan",
             "kiyoda",
             "imamura",
             "hatsune",
             "aozora",
             ]
```

```python
tagger = MeCab.Tagger("-Ochasen -u /Users/yuki/foo/bar/user_dic2.dic")
tagger.parse("")
```

```python
def tokenize_ja(text):
    node = tagger.parseToNode(str(text)) #分かち書きしたのが入っている。イテレータ
    while node:
        surface = node.surface.lower() #surfaceは単語の見た目　例：トろ
        hinshi = node.feature.split(",")[0]
        if surface != "" and hinshi != "記号":
            yield surface
        node = node.next
```

```python
def tokenize(content):
    return [token for token in tokenize_ja(content)]
```

```python
df = pd.read_csv("../pretest_mitani_lunch.csv")
```

```python
for i, path in enumerate(lunch_file):
    df = pd.read_csv(path)
    l = []
    for d in df["lunch_review"]:
        l.append("/".join(tokenize(d)))
    data = pd.DataFrame(l, columns = ["分ち書き"])
    data.to_csv(store_name[i] + "_wakati_lunch.csv", index=False)
```

```python
ll = []
for i, path in enumerate(dinner_file):
    df = pd.read_csv(path)
    l = []
    for d in df["dinner_review"]:
        l.append("/".join(tokenize(d)))
    data = pd.DataFrame(l, columns = ["分ち書き"])
    data.to_csv(store_name[i] + "_wakati_dinner.csv", index=False)
    ll.append(l)
```
