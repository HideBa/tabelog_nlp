```python
stop_words = ["\u25c9", "\u2764", "\ufe0e", "\u9ad9", "\uff5e", "\uff5e"]
```

```python
def tokenize_ja(text):
    node = tagger.parseToNode(str(text)) #分かち書きしたのが入っている。イテレータ
    while node:
        surface = node.surface.lower() #surfaceは単語の見た目　例：トろ
        hinshi = node.feature.split(",")[0]
        if "\\u" not in surface:
            yield surface
        node = node.next
```

```python
def tokenize(content):
    return [token for token in tokenize_ja(content)]
```

```python
l = []
for i, path in enumerate(lunch_file):
    df = pd.read_csv(path)
    ll = []
    for d in df["lunch_review"]:
        
        for sentence in re.split("[。！!？… \. \?]", d):
            ll.append("".join(tokenize(sentence)))
    l.append(ll)
    data = pd.DataFrame(ll, columns = ["分ち書き"])
    data.to_csv(store_name[i] + "_wakati_lunch.csv")
```
