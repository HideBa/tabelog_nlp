```python
store_name = ["sushitsuu",
             "sushiichi",
             "sawada",
             "mitani",
             "matsukan",
             "kiyoda",
             "imamura",
             "hatsune",
             "aozora",
             ]
```

```python
stop_words = ["♦","▪","⚪","・","◾",]
```

```python
def tokenize_ja(text):
    node = tagger.parseToNode(str(text)) #分かち書きしたのが入っている。イテレータ
    while node:
        surface = node.surface.lower() #surfaceは単語の見た目　例：トろ
        feature = node.feature.split(",")
        hinshi = feature[0]
        genkei = feature[6]
        if surface not in stop_words and surface != "" and hinshi in ["形容詞", "動詞", "副詞", "助動詞", "接続詞", "連体詞", "感動詞", "接頭詞"]:
            yield genkei
        elif surface not in stop_words and surface != "" and hinshi == "名詞":
            yield surface
        node = node.next
```

```python
def tokenize(content):
    return [token for token in tokenize_ja(content)]
```

```python
def count_word(file):
    dic = defaultdict(int)
    df = pd.read_csv(file)
    if "dinner" in file:
        for review in df["dinner_review"]:
            for sentence in review.split("。"):
                t = tokenize(sentence)
                for tt in t:
                    dic[tt] += 1
        counter = Counter.most_common(dic)
        return counter
    else:
        for review in df["lunch_review"]:
            for sentence in review.split("。"):
                t = tokenize(sentence)
                for tt in t:
                    dic[tt] += 1
        counter = Counter.most_common(dic)
    return counter
```

```python
for i, file in enumerate(lunch_file):
    dic = count_word(file)
    name = store_name[i]
    counter = count_word(file)
    l = []
    for c in counter:
        l.append([c[0], c[1]])
    df = pd.DataFrame(l, columns=["単語","登場回数"])
    df.to_csv(name+"_frequent_lunch.csv")
```
